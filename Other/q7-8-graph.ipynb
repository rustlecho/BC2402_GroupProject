```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data
data = [
    ['Qatar Airways', 'Business', 'FoodnBeverages', 16, 16, 80, 14.29, 14.29, 71.43],
    ['Qatar Airways', 'Business', 'InflightEntertainment', 8, 10, 94, 7.14, 8.93, 83.93],
    ['Qatar Airways', 'Business', 'SeatComfort', 15, 12, 85, 13.39, 10.71, 75.89],
    ['Qatar Airways', 'Business', 'StaffService', 18, 8, 86, 16.07, 7.14, 76.79],
    ['Qatar Airways', 'Business', 'ValueForMoney', 28, 14, 70, 25.00, 12.50, 62.50],
    ['Qatar Airways', 'Couple Leisure', 'FoodnBeverages', 17, 15, 98, 13.08, 11.54, 75.38],
    ['Qatar Airways', 'Couple Leisure', 'InflightEntertainment', 15, 18, 97, 11.54, 13.85, 74.62],
    ['Qatar Airways', 'Couple Leisure', 'SeatComfort', 27, 13, 90, 20.77, 10.00, 69.23],
    ['Qatar Airways', 'Couple Leisure', 'StaffService', 13, 8, 109, 10.00, 6.15, 83.85],
    ['Qatar Airways', 'Couple Leisure', 'ValueForMoney', 40, 15, 75, 30.77, 11.54, 57.69],
    ['Qatar Airways', 'Family Leisure', 'FoodnBeverages', 25, 8, 79, 22.32, 7.14, 70.54],
    ['Qatar Airways', 'Family Leisure', 'InflightEntertainment', 3, 10, 99, 2.68, 8.93, 88.39],
    ['Qatar Airways', 'Family Leisure', 'SeatComfort', 21, 17, 74, 18.75, 15.18, 66.07],
    ['Qatar Airways', 'Family Leisure', 'StaffService', 14, 12, 86, 12.50, 10.71, 76.79],
    ['Qatar Airways', 'Family Leisure', 'ValueForMoney', 25, 19, 68, 22.32, 16.96, 60.71],
    ['Qatar Airways', 'Solo Leisure', 'FoodnBeverages', 41, 36, 245, 12.73, 11.18, 76.09],
    ['Qatar Airways', 'Solo Leisure', 'InflightEntertainment', 24, 34, 264, 7.45, 10.56, 81.99],
    ['Qatar Airways', 'Solo Leisure', 'SeatComfort', 33, 42, 247, 10.25, 13.04, 76.71],
    ['Qatar Airways', 'Solo Leisure', 'StaffService', 28, 40, 254, 8.70, 12.42, 78.88],
    ['Qatar Airways', 'Solo Leisure', 'ValueForMoney', 57, 39, 226, 17.70, 12.11, 70.19],
    ['Singapore Airlines', 'Business', 'FoodnBeverages', 31, 23, 89, 21.68, 16.08, 62.24],
    ['Singapore Airlines', 'Business', 'InflightEntertainment', 17, 18, 108, 11.89, 12.59, 75.52],
    ['Singapore Airlines', 'Business', 'SeatComfort', 26, 21, 96, 18.18, 14.69, 67.13],
    ['Singapore Airlines', 'Business', 'StaffService', 17, 20, 106, 11.89, 13.99, 74.13],
    ['Singapore Airlines', 'Business', 'ValueForMoney', 45, 23, 75, 31.47, 16.08, 52.45],
    ['Singapore Airlines', 'Couple Leisure', 'FoodnBeverages', 74, 42, 132, 29.84, 16.94, 53.23],
    ['Singapore Airlines', 'Couple Leisure', 'InflightEntertainment', 31, 34, 183, 12.50, 13.71, 73.79],
    ['Singapore Airlines', 'Couple Leisure', 'SeatComfort', 58, 43, 147, 23.39, 17.34, 59.27],
    ['Singapore Airlines', 'Couple Leisure', 'StaffService', 52, 30, 166, 20.97, 12.10, 66.94],
    ['Singapore Airlines', 'Couple Leisure', 'ValueForMoney', 79, 36, 133, 31.85, 14.52, 53.63],
    ['Singapore Airlines', 'Family Leisure', 'FoodnBeverages', 42, 29, 111, 23.08, 15.93, 60.99],
    ['Singapore Airlines', 'Family Leisure', 'InflightEntertainment', 27, 27, 128, 14.84, 14.84, 70.33],
    ['Singapore Airlines', 'Family Leisure', 'SeatComfort', 31, 32, 119, 17.03, 17.58, 65.38],
    ['Singapore Airlines', 'Family Leisure', 'StaffService', 31, 28, 123, 17.03, 15.38, 67.58],
    ['Singapore Airlines', 'Family Leisure', 'ValueForMoney', 51, 21, 110, 28.02, 11.54, 60.44],
    ['Singapore Airlines', 'Solo Leisure', 'FoodnBeverages', 82, 51, 197, 24.85, 15.45, 59.70],
    ['Singapore Airlines', 'Solo Leisure', 'InflightEntertainment', 43, 52, 235, 13.03, 15.76, 71.21],
    ['Singapore Airlines', 'Solo Leisure', 'SeatComfort', 63, 48, 219, 19.09, 14.55, 66.36],
    ['Singapore Airlines', 'Solo Leisure', 'StaffService', 54, 41, 235, 16.36, 12.42, 71.21],
    ['Singapore Airlines', 'Solo Leisure', 'ValueForMoney', 73, 38, 219, 22.12, 11.52, 66.36]
]

# Convert data to pandas DataFrame
columns = ['Airline', 'TypeofTraveller', 'RatingType', 'Negative', 'Neutral', 'Positive', 
           'NegativePercentage', 'NeutralPercentage', 'PositivePercentage']
df = pd.DataFrame(data, columns=columns)

# Function to create separate plots for each type of traveller
def separated_stacked_bar(dataframe, airline):
    airline_data = dataframe[dataframe['Airline'] == airline]
    traveller_types = airline_data['TypeofTraveller'].unique()
    rating_types = airline_data['RatingType'].unique()

    for traveller in traveller_types:
        traveller_data = airline_data[airline_data['TypeofTraveller'] == traveller]
        negative = traveller_data['NegativePercentage'].values
        neutral = traveller_data['NeutralPercentage'].values
        positive = traveller_data['PositivePercentage'].values

        fig, ax = plt.subplots(figsize=(10, 6))
        x = np.arange(len(rating_types))  # Base x positions
        category_colors = ['salmon', 'silver', 'lightgreen']

        ax.bar(x, negative, 0.4, label="Negative", color=category_colors[0])
        ax.bar(x, neutral, 0.4, bottom=negative, label="Neutral", color=category_colors[1])
        ax.bar(x, positive, 0.4, bottom=negative + neutral, label="Positive", color=category_colors[2])

        # Add annotations for percentages
        for i, (neg, neu, pos) in enumerate(zip(negative, neutral, positive)):
            if neg > 0:
                ax.text(i, neg / 2, f"{neg:.1f}%", ha='center', va='center', fontsize=8, color='black')
            if neu > 0:
                ax.text(i, neg + neu / 2, f"{neu:.1f}%", ha='center', va='center', fontsize=8, color='black')
            if pos > 0:
                ax.text(i, neg + neu + pos / 2, f"{pos:.1f}%", ha='center', va='center', fontsize=8, color='black')

        # Customize plot
        ax.set_title(f"Ratings for {traveller} Travellers - {airline}", fontsize=16)
        ax.set_xlabel("Rating Types", fontsize=12)
        ax.set_ylabel("Percentage (%)", fontsize=12)
        ax.set_xticks(x)
        ax.set_xticklabels(rating_types, fontsize=10)
        ax.legend(title="Rating Category", fontsize=10)
        ax.set_ylim(0, 100)

        plt.tight_layout()
        plt.show()
```

```python
def plot_grouped_bar_chart(dataframe, title):
    categories = ["Avg(SeatComfort)", "Avg(FoodnBeverages)", "Avg(InflightEntertainment)", 
                  "Avg(ValueForMoney)", "Avg(OverallRating)"]
    airlines = dataframe["Airline"].unique()
    colors = ["lightblue", "bisque", "plum", "salmon"]  # Custom colors for each class

    for airline in airlines:
        airline_data = dataframe[dataframe["Airline"] == airline]
        class_labels = airline_data["Class"].astype(str)
        x = range(len(categories))
        
        plt.figure(figsize=(12, 8))
        
        # Plot grouped bars for each class within the airline
        for idx, airline_class in enumerate(class_labels.unique()):
            class_data = airline_data[airline_data["Class"] == airline_class]
            means = class_data[categories].mean()
            bar_positions = [p + idx * 0.15 for p in x]
            plt.bar(bar_positions, means, width=0.15, label=airline_class, color=colors[idx % len(colors)])
            
            # Annotate each bar with the rating value
            for pos, value in zip(bar_positions, means):
                plt.text(pos, value + 0.1, f"{value:.2f}", ha='center', fontsize=9)
        
        plt.title(f"{title} - {airline}")
        plt.xlabel('Rating Categories')
        plt.ylabel('Average Ratings')
        plt.xticks([p + 0.225 for p in x], categories)
        plt.legend(title="Class")
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

plot_grouped_bar_chart(table_7_1_df_reordered, "Average Ratings for Seasonal Period")
plot_grouped_bar_chart(table_7_2_df_reordered, "Average Ratings for Non-Seasonal Period")
```
